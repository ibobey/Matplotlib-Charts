```python
# Import Necessary Modules
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from numpy import pi
```

__Radar chart__ as known spider-charts is a kind of __polar chart__ . Using radar charts __provides us to varios benefits__ like comparing data multi-dimensionly, showing weak and robust features of data  etc.

### Basics of Radar Chart

In this sample chart we will plot 5 score about related categories.

```python
# Create Test Data (5 mark)
test_data = np.random.randint(1,6,5)
```

Now, We need to add __first element__ of array into the __end__ of list. If we don't add first element technically nothing is changed but the aim of adding first element to end is getting __closed radar chart__.

```python
# Add first element to end of list 
test_data = np.concatenate((test_data,[test_data[0]]))
```

```python
test_data
```

Instead of using __linear__ scales, we should use __radianal__ scale. Radianal scales makes easier to determine __polar__ chart coordinates

```python
label_placement = np.linspace(0,2*pi,num=len(test_data)) # radian
```

* num=len(test_data) makes us to 5+1 equal radar region to plot our data in it.

```python
# 0 radian to 2*pi radian
label_placement 
```

```python
categories = ['A', 'B', 'C', 'D', 'E',""]
```

* In the begining we told, We have five categories (A-E) to visualize but what is __blank__ category __" "__ ?
* As the mentioned at the begining, to __make closed radar chart__ instead of polar __opened lines__ we have to to define a blank(non-visible) category.

```python
# Setting Seaborn themes
sns.set_theme(style="ticks", palette="pastel")
sns.set_context("paper")
```

Now we are ready to plotting. Then we will inspect the code

```python
plt.figure(figsize=(6,6)) # Creating Figure Object
plt.subplot(polar=True) # We need to define polar plot 

plt.ylim(0,5) # To get make aesthetics plot ve set y-limitors

plt.plot(label_placement, #x data 
         test_data, # y_data
         label="test data", 
         color="#009B77")


plt.thetagrids(np.degrees(label_placement), labels=categories) # Creating polar spider - lines 


# To make solid area we need to fill closed area 
plt.fill_between(label_placement,
                 test_data,
                 0,
                 color="#009B77",
                 alpha=0.1)


plt.grid(True,alpha=0.5,ls="--")
plt.legend(loc='best')
plt.title("TEST DATA")
plt.show()
```

Let's Inspect the code:
* First we created figure object
* Then we setted a y-limitor because of we __know__ our __min__ value is 1 and __max__ is 5. The aim of setting y-lims is make plot more __unterstandable__ and __aesthetics__.
* Then we plotted our data like a normal line-chart. In this step, We used a __radianal__ x_data, beceuse we are working on __polar coordinate system__ and __radianal__ scale makes __plotting__ and __tracing__ __eaiser__ .
* After, we created thetagrids.
* Then, We filled closed area. This step, using radianal scale made easier to __determinig x,y and base line__ to fill area

###  Make Multiple Plots

```python
marks_1 = np.random.randint(40,100,5)
marks_1 = np.concatenate((marks_1,[marks_1[0]]))

marks_2 = np.random.randint(40,100,5)
marks_2 = np.concatenate((marks_2,[marks_2[0]]))

marks_3 = np.random.randint(40,100,5)
marks_3 = np.concatenate((marks_3,[marks_3[0]]))
```

```python
categories = ['Math', 'F.L.', 'P.E.', 'BIO', 'CHE', '']
```

```python
label_placement = np.linspace(0,2*pi,num=len(marks_1)) # radian
```

__Caution__: We need to watch lenght of arrays equality

```python
sns.set_theme(style="ticks", palette="pastel")
sns.set_context("paper")
```

```python
plt.figure(figsize=(6,6))
plt.subplot(polar=True)

plt.ylim(0,100)

plt.plot(label_placement, 
         marks_1,
         label="Student 1",
         color="#009B77")

plt.plot(label_placement, 
         marks_2,
         label="Student 2",
         color="#DD4124")

plt.plot(label_placement, 
         marks_3,
         label="Student 3",
         color="#EFC050")


plt.thetagrids(np.degrees(label_placement), labels=categories)


plt.grid(True,alpha=0.5,ls="-.")
plt.legend(loc='best')
plt.tight_layout()
plt.title("QUIZ MARKS")
plt.show()
```

### NOTES

Making Radar-Chart is easy when you watch the main rules;
* Use radioanal scale to determine x-y coordinates
* Add a blank and non-visual value and category to get closed chart and closed chart makes easier to make solid filled area.
* Use y- limitors according to max & min values of data
* Using specific score-grid for y-lims makes the chart more easthetics
* Using Nan or Null values without normalization could ruin of radar chart anatomy
